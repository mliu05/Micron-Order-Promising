```python
machine_capacity_assembly = machine_capacity.drop(columns=['Testing_Machine']).dropna().reset_index()
machine_capacity_testing = machine_capacity.drop(columns=['Assembly_Machine']).dropna().reset_index()

ind_dict = [0]*(machine_capacity_assembly.shape[0])
val_dict =[0]*(machine_capacity_assembly.shape[0])
for i in np.arange(0,machine_capacity_assembly.shape[0]):
    val_dict[i] = float(machine_capacity_assembly.Hours_Available[i])
    ind_dict[i] = tuple([machine_capacity_assembly.Factory[i],machine_capacity_assembly.Assembly_Machine[i],machine_capacity_assembly.Week[i]])
CAA = dict(zip(ind_dict,val_dict))

ind_dict = [0]*(machine_capacity_testing.shape[0])
val_dict =[0]*(machine_capacity_testing.shape[0])
for i in np.arange(0,machine_capacity_testing.shape[0]):
    val_dict[i] = float(machine_capacity_testing.Hours_Available[i])
    ind_dict[i] = tuple([machine_capacity_testing.Factory[i],machine_capacity_testing.Testing_Machine[i],machine_capacity_testing.Week[i]])
CAT = dict(zip(ind_dict,val_dict))
```

# Create Decision Variables

```python
# Create variables as dictionaries
X = LpVariable.dicts(
    "chips_to_assem", i_c_d_f_w_1, lowBound=0, cat="Continuous")
Y = LpVariable.dicts(
    "chips_to_test",i_c_d_f_w_1,lowBound=0,cat="Continuous")

XM = LpVariable.dicts(
    "mods_to_assem", m_c_d_f_w_1, lowBound=0, cat="Continuous")
YM = LpVariable.dicts(
    "mods_to_test",m_c_d_f_w_1,lowBound=0,cat="Continuous")

U = LpVariable.dicts(
    "unfulfilled_chips", i_c_d_w, lowBound=0, cat="Continuous")
UM = LpVariable.dicts(
    "unfulfilled_mods",m_c_d_w,lowBound=0,cat="Continuous")

BO = LpVariable.dicts(
    "backordered_chips", i_c_d_w, lowBound=0, cat="Continuous")
BOM = LpVariable.dicts(
    "backordered_mods",m_c_d_w,lowBound=0,cat="Continuous")
ABO = LpVariable.dicts(
    "actual_backordered_chips", i_c_d_w, lowBound=0, cat="Continuous")
ABOM = LpVariable.dicts(
    "actual_backordered_mods",m_c_d_w,lowBound=0,cat="Continuous")

LS = LpVariable.dicts(
    "lost_sale_chips", i_c_d_w, lowBound=0, cat="Continuous")
LSM = LpVariable.dicts(
    "lost_sale_mods",m_c_d_w,lowBound=0,cat="Continuous")
ALS = LpVariable.dicts(
    "actual_lost_sale_chips", i_c_d_w, lowBound=0, cat="Continuous")
ALSM = LpVariable.dicts(
    "actual_lost_sale_mods",m_c_d_w,lowBound=0,cat="Continuous")

ITC = LpVariable.dicts(
    "tested_chip_inv", i_c_d_f_w_0, lowBound=0, cat="Continuous")
ITMC = LpVariable.dicts(
    "tested_mod_inv",m_c_d_f_w_0,lowBound=0,cat="Continuous")
IDC = LpVariable.dicts(
    "die_inv", i_c_d_f_w_0, lowBound=0, cat="Continuous")

IUC = LpVariable.dicts(
    "untested_chip_inv", i_c_d_f_w_0, lowBound=0, cat="Continuous")
IUMC = LpVariable.dicts(
    "untested_mod_inv",m_c_d_f_w_0,lowBound=0,cat="Continuous")

Q = LpVariable.dicts(
    "shipped_chips", i_c_d_f_w_1, lowBound=0, cat="Continuous")
QM = LpVariable.dicts(
    "shipped_mods",i_c_d_f_w_1,lowBound=0,cat="Continuous")
```

# Initialize LP Model

```python
#initialize model/problem
problem= LpProblem(name="problem", sense=LpMinimize)
```

# Objective Function

```python
#objective function

problem += (lpSum([L[(i,c)] * ABO[(i,c,d,w)] for (i,c,d,w) in i_c_d_w_1]) 
            + lpSum([L[(m,c)] * ABOM[(m,c,d,w)] for (m,c,d,w) in m_c_d_w_1]) 
            + lpSum([H[(i)] * IUC[(i,c,d,f,w)] + ITC[(i,c,d,f,w)] for (i,c,d,f,w) in i_c_d_f_w_1]) 
            + lpSum([HM[(m)] * IUMC[(m,c,d,f,w)] + ITMC[(m,c,d,f,w)] for (m,c,d,f,w) in m_c_d_f_w_1]) 
            + lpSum([ HD[(i)] * IDC[(i,c,d,f,w)] for (i,c,d,f,w) in i_c_d_f_w_1]) 
            + lpSum([ P[(i)] * ALS[(i,c,d,w)] for (i,c,d,w) in i_c_d_w_1]) 
            +lpSum([ PM[(m)] * ALSM[(m,c,d,w)] for (m,c,d,w) in m_c_d_w_1]) 
            + lpSum([ CC[(f,s)] * X[(i,c,d,f,1)] * PTA[(i,f,s)] for (i,c,d,f,s) in i_c_d_f_s]) 
            + lpSum([ CC[(f,s)] * XM[(m,c,d,f,1)] * PTAM[(m,f,s)] for (m,c,d,f,s) in m_c_d_f_s])     
            + lpSum([ (CC[(f,s)] + MC[(s,w-1)] + ep) * X[(i,c,d,f,w)] * PTA[(i,f,s)] for (i,c,d,f,s,w) in i_c_d_f_s_w2]) 
            + lpSum([ (CC[(f,s)]+ MC[(s,w-1)] + ep) * XM[(m,c,d,f,w)] * PTAM[(m,f,s)] for (m,c,d,f,s,w) in m_c_d_f_s_w2]) 
            + lpSum([ CCT[(f,t)] * Y[(i,c,d,f,1)] * PTT[(i,f,t)] for (i,c,d,f,t) in i_c_d_f_t]) 
            + lpSum([ CCT[(f,t)] * YM[(m,c,d,f,1)] * PTTM[(m,f,t)] for (m,c,d,f,t) in m_c_d_f_t])  
            + lpSum([ (CCT[(f,t)] + MCT[(t,w-1)] + ep) * Y[(i,c,d,f,w)] * PTT[(i,f,t)] for (i,c,d,f,t,w) in i_c_d_f_t_w2]) 
            + lpSum([ (CCT[(f,t)]+ MCT[(t,w-1)] + ep) * YM[(m,c,d,f,w)] * PTTM[(m,f,t)] for (m,c,d,f,t,w) in m_c_d_f_t_w2]) , "Total Cost") 
```

# Constraints

```python
#die bank inv
for (chip,fac) in i_f:
    problem += lpSum([IDC[(i,c,d,f,0)] for (i,c,d,f) in i_c_d_f if i == chip and f==fac]) == I0[(chip,fac)], ("Starting_Die_Inv" + str((chip,fac)))
for (i,c,d,f,w) in i_c_d_f_w_1:
    problem += IDC[(i,c,d,f,w)] == IDC[(i,c,d,f,w-1)] - X[(i,c,d,f,w)] + INC[(i,c,d,f,w)], ("Die_Inv" + str((i,c,d,f,w)))
```

```python
#untested assemblies inv
for (i,c,d,f,w) in i_c_d_f_w_1:
    problem += IUC[(i,c,d,f,w)] == IUC[(i,c,d,f,w-1)] + X[(i,c,d,f,w)] - Y[(i,c,d,f,w)], ("Untested_chip_inv" + str((i,c,d,f,w)))
for (chip,fac) in i_f:
    problem += lpSum([IUC[(i,c,d,f,0)] for (i,c,d,f) in i_c_d_f if i == chip and f==fac]) == IU0[(chip,fac)], ("Starting_untested_chip_Inv" + str((chip,fac)))

for (m,c,d,f,w) in m_c_d_f_w_1:
    problem += IUMC[(m,c,d,f,w)] == IUMC[(m,c,d,f,w-1)] + XM[(m,c,d,f,w)] - YM[(m,c,d,f,w)], ("Untested_mod_inv" + str((m,c,d,f,w)))
for (mod,fac) in m_f:
    problem += lpSum([IUMC[(m,c,d,f,0)] for (m,c,d,f) in m_c_d_f if m == mod and f==fac]) == IUM0[(mod,fac)], ("Starting_untested_mod_Inv" + str((mod,fac)))
```

```python
#tested assemblies inv
for (chip,fac) in i_f:
    problem += lpSum([ITC[(i,c,d,f,0)] for (i,c,d,f) in i_c_d_f if i == chip and f==fac]) == IT0[(chip,fac)], ("Starting_tested_chip_Inv" + str((chip,fac)))
for (i,c,d,f,w) in i_c_d_f_w_1:
    problem += ITC[(i,c,d,f,w)] == ITC[(i,c,d,f,w-1)] + Y[(i,c,d,f,w)] - Q[(i,c,d,f,w)] - lpSum([R[(i,m)] * XM[(m,c,d,f,w)] for m in np.arange(1,M+1)]), ("Tested_chip_inv" + str((i,c,d,f,w)))

    
for (mod,fac) in m_f:
    problem += lpSum([ITMC[(m,c,d,f,0)] for (m,c,d,f) in m_c_d_f if m == mod and f==fac]) == ITM0[(mod,fac)], ("Starting_tested_mod_Inv" + str((mod,fac)))
for (m,c,d,f,w) in m_c_d_f_w_1:
    problem += ITMC[(m,c,d,f,w)] == ITMC[(m,c,d,f,w-1)] + YM[(m,c,d,f,w)] - QM[(m,c,d,f,w)], ("Tested_mod_inv" + str((m,c,d,f,w)))
```

```python
#supply balance
for (i,c,d,f,w) in i_c_d_f_w_1:
    problem += X[(i,c,d,f,w)] <= IDC[(i,c,d,f,w-1)] + INC[(i,c,d,f,w)], ("Chip_Assem_Supply" + str((i,c,d,f,w)))
    problem += Y[(i,c,d,f,w)] <= IUC[(i,c,d,f,w-1)] + X[(i,c,d,f,w)], ("Chip_Test_Supply" + str((i,c,d,f,w)))
    problem += lpSum([XM[(m,c,d,f,w)]*R[(i,m)] for m in np.arange(1,M+1)]) <= ITC[(i,c,d,f,w-1)] + Y[(i,c,d,f,w)] - Q[(i,c,d,f,w)], ("Mod_Assem_Supply" + str((i,c,d,f,w)))
    
    
for (i,c,d,f,w) in i_c_d_f_w_1:
    problem += Q[(i,c,d,f,w)] <= ITC[(i,c,d,f,w-1)] + Y[(i,c,d,f,w)] - lpSum([R[(i,m)] * XM[(m,c,d,f,w)] for m in np.arange(1,M+1)]),("Chips_Avail_to_Ship" + str((i,c,d,f,w)))
for (m,c,d,f,w) in m_c_d_f_w_1:
    problem += QM[(m,c,d,f,w)] <= ITMC[(m,c,d,f,w-1)] + YM[(m,c,d,f,w)],("Mods_Avail_to_Ship" + str((m,c,d,f,w)))
    problem += YM[(m,c,d,f,w)] <= IUMC[(m,c,d,f,w-1)] + XM[(m,c,d,f,w)], ("Mod_Test_Supply" + str((m,c,d,f,w)))
```

```python
#SHIPPED/ALLOCATED QUANTITY

for w in np.arange(1,W+1):
    for (i,c,d,wi) in i_c_d_w_1:
        if wi == w & wi<d:
            problem += lpSum([ Q[(i,c,d,f,wi)] for f in np.arange(1,F+1)]) >= DI[(i,c,d,w)] - U[(i,c,d,wi)],("Shipped_Chip" + str((i,c,d,w,wi)))
        elif wi>w & wi<d:
            problem += lpSum([ Q[(i,c,d,f,wi)] for f in np.arange(1,F+1)]) >= U[(i,c,d,wi-1)] - U[(i,c,d,wi)],("Shipped_Chip" + str((i,c,d,w,wi)))
        elif wi>w & wi==d:
            problem += lpSum([ Q[(i,c,d,f,wi)] for f in np.arange(1,F+1)]) >= U[(i,c,d,wi-1)] - BO[(i,c,d,wi)],("Shipped_Chip" + str((i,c,d,w,wi)))
        elif wi>w & wi>d & wi<d+sigmai[(i,c)]:
            problem += lpSum([ Q[(i,c,d,f,wi)] for f in np.arange(1,F+1)]) >= BO[(i,c,d,wi-1)] - BO[(i,c,d,wi)],("Shipped_Chip" + str((i,c,d,w,wi)))
        else:
            problem += lpSum([ Q[(i,c,d,f,wi)] for f in np.arange(1,F+1)]) >= 0,("Shipped_Chip" + str((i,c,d,w,wi)))

for w in np.arange(1,W+1):
    for (m,c,d,wi) in m_c_d_w_1:
        if wi == w & wi<d:
            problem += lpSum([ QM[(m,c,d,f,wi)] for f in np.arange(1,F+1)]) >= DM[(m,c,d,w)] - UM[(m,c,d,wi)],("Shipped_Mod" + str((m,c,d,w,wi)))
        elif wi>w & wi<d:
            problem += lpSum([ QM[(m,c,d,f,wi)] for f in np.arange(1,F+1)]) >= UM[(m,c,d,wi-1)] - UM[(m,c,d,wi)],("Shipped_Mod" + str((m,c,d,w,wi)))
        elif wi>w & wi==d:
            problem += lpSum([ QM[(m,c,d,f,wi)] for f in np.arange(1,F+1)]) >= UM[(m,c,d,wi-1)] - BOM[(m,c,d,wi)],("Shipped_Mod" + str((m,c,d,w,wi)))
        elif wi>w & wi>d & wi<d+sigmam[m,c]:
            problem += lpSum([ QM[(m,c,d,f,wi)] for f in np.arange(1,F+1)]) >= BOM[(m,c,d,wi-1)] - BOM[(m,c,d,wi)],("Shipped_Mod" + str((m,c,d,w,wi)))
        else:
            problem += lpSum([ QM[(m,c,d,f,wi)] for f in np.arange(1,F+1)]) >= 0,("Shipped_Mod" + str((m,c,d,w,wi)))
```

```python
#BACKORDERING

for w in np.arange(1,W+1):
    for (i,c,d,wi) in i_c_d_w_1: 
        if wi == w:
            problem += U[(i,c,d,wi)] >= DI[(i,c,d,w)] - lpSum([(ITC[(i,c,d,f,wi-1)] - Y[(i,c,d,f,wi)]) for f in np.arange(1,F+1)]) + lpSum([R[i,m]*XM[m,c,d,f,wi] for (m,f) in m_f]), ("Unfulfilled_Chips" + str((i,c,d,w,wi)))
        elif wi>w and wi<d:
            problem += U[(i,c,d,wi)] >= U[(i,c,d,wi-1)] - lpSum([(ITC[(i,c,d,f,wi-1)] - Y[(i,c,d,f,wi)]) for f in np.arange(1,F+1)]) + lpSum([R[i,m]*XM[m,c,d,f,wi] for (m,f) in m_f]), ("Unfulfilled_Chips" + str((i,c,d,w,wi)))
        else:
            problem += U[(i,c,d,wi)] >= 0 , ("Unfulfilled_Chips" + str((i,c,d,w,wi)))

for w in np.arange(1,W+1):
    for (m,c,d,wi) in m_c_d_w_1:
        if wi == w:
            problem += UM[(m,c,d,wi)] >= DM[(m,c,d,w)] - lpSum([(ITMC[(m,c,d,f,wi-1)] - YM[(m,c,d,f,wi)]) for f in np.arange(1,F+1)]), ("Unpenalized_Mod" + str((m,c,d,w,wi)))
        elif w>d and wi<d:
            problem += UM[(m,c,d,wi)] >= UM[(m,c,d,wi-1)] - lpSum([(ITMC[(m,c,d,f,wi-1)] - YM[(m,c,d,f,wi)]) for f in np.arange(1,F+1)]), ("Unpenalized_Mod" + str((m,c,d,w,wi))) 
        else:
            problem += UM[(m,c,d,wi)] >= 0, ("Unfulfilled_mod" + str((m,c,d,w,wi)))         
            
            
            

for (i,c,d,w) in i_c_d_w_1:
    if w==d:
        problem += BO[(i,c,d,w)] >= U[(i,c,d,w-1)] - lpSum([(ITC[(i,c,d,f,w-1)] - Y[(i,c,d,f,w)]) for f in np.arange(1,F+1)]) + lpSum([R[i,m]*XM[m,c,d,f,w] for (m,f) in m_f]), ("Backorder_chip" + str((i,c,d,w)))
    elif w>d and w<d+sigmai[i,c]:
        problem += BO[(i,c,d,w)] >= BO[(i,c,d,w-1)] - lpSum([(ITC[(i,c,d,f,w-1)] - Y[(i,c,d,f,w)]) for f in np.arange(1,F+1)]) + lpSum([R[i,m]*XM[m,c,d,f,w] for (m,f) in m_f]), ("Backorder_chip" + str((i,c,d,w)))
    else:
        problem += BO[(i,c,d,w)] >= 0, ("Backorder_chip" + str((i,c,d,w))) 

for (i,c,d,w) in i_c_d_w_1:
    if w==d:
        problem += BO[(i,c,d,w)] >= DI[i,c,d,w]- lpSum([(ITC[(i,c,d,f,w-1)] - Y[(i,c,d,f,w)]) for f in np.arange(1,F+1)]) + lpSum([R[i,m]*XM[m,c,d,f,w] for (m,f) in m_f]), ("Backorder_Chip_SameDueOrder" + str((i,c,d,w)))
    else:
        problem += BO[(i,c,d,w)] >= 0, ("Backorder_Chip_SameDueOrder" + str((i,c,d,w)))

                                                                                                                                 
for (m,c,d,w) in m_c_d_w_1:
    if w==d:
        problem += BOM[(m,c,d,w)] >= UM[(m,c,d,w-1)] - lpSum([(ITMC[(m,c,d,f,w-1)] - YM[(m,c,d,f,w)]) for f in np.arange(1,F+1)]), ("Backorder_mod" + str((m,c,d,w)))
    elif w>d and w<d+sigmam[m,c]:
        problem += BOM[(m,c,d,w)] >= BOM[(m,c,d,w-1)] - lpSum([(ITMC[(m,c,d,f,w-1)] - YM[(m,c,d,f,w)]) for f in np.arange(1,F+1)]), ("Backorder_mod" + str((m,c,d,w))) 
    else:
        problem += BOM[(m,c,d,w)] >= 0, ("Backorder_mod" + str((m,c,d,w)))

for (m,c,d,w) in m_c_d_w_1:
    if w==d:
        problem += BOM[(m,c,d,w)] >= DM[(m,c,d,w)]- lpSum([(ITMC[(m,c,d,f,w-1)] - YM[(m,c,d,f,w)]) for f in np.arange(1,F+1)]), ("Backorder_Mod_SameDueOrder" + str((m,c,d,w)))
    else:
        problem += BOM[(m,c,d,w)] >= 0, ("Backorder_Mod_SameDueOrder" + str((m,c,d,w)))
    
```

```python
#LOST SALES

for (i,c,d,w) in i_c_d_w_1:
    if w == d+sigmai[(i,c)]:
        problem += LS[(i,c,d,w)] >= BO[(i,c,d,w-1)] - lpSum([ITC[(i,c,d,f,w-1)] - Y[(i,c,d,f,w)] for f in np.arange(1,F+1)]) + lpSum([R[i,m]*XM[m,c,d,f,w] for (m,f) in m_f]), ("Dummy_lost_Sales_Chip" + str((i,c,d,w)))
    else:
        problem += LS[(i,c,d,w)] >= 0, ("Dummy_lost_Sales_Chip" + str((i,c,d,w)))
        
for (m,c,d,w) in m_c_d_w_1:
    if w==d+sigmam[(m,c)]:
        problem += LSM[(m,c,d,w)] >= BOM[(m,c,d,w-1)] - lpSum([ITMC[(m,c,d,f,w-1)] - YM[(m,c,d,f,w)] for f in np.arange(1,F+1)]), ("Dummy_lost_Sales_Mod" + str((m,c,d,w)))
    else:
        problem += LSM[(m,c,d,w)] >= 0, ("Dummy_lost_Sales_Mod" + str((m,c,d,w)))
        
        
for (i,c,d,w) in i_c_d_w_1:
    if w==d and d+sigmai[(i,c)] <= W:
        ALS[(i,c,d,w)] >= LS[(i,c,d,d+sigmai[(i,c)])], ("Actual_lost_Sales_Chip" + str((i,c,d,w)))
    else:
        ALS[(i,c,d,w)] >= 0, ("Actual_lost_Sales_Chip" + str((i,c,d,w)))

for (m,c,d,w) in m_c_d_w_1:
    if w==d and d+sigmam[(m,c)] <= W:
        ALSM[(m,c,d,w)] >= LSM[(m,c,d,d+sigmam[(m,c)])], ("Actual_lost_Sales_Mod" + str((m,c,d,w)))
    else:
        ALSM[(m,c,d,w)] >= 0, ("Actual_lost_Sales_Mod" + str((m,c,d,w)))
```

```python
#updating BOs
for (i,c,d,w) in i_c_d_w_1:
    if w>=d and w<d+sigmai[(i,c)]:
        ABO[(i,c,d,w)] >= BO[(i,c,d,w)] - ALS[(i,c,d,d)],("Actual_BO_Chip" + str((i,c,d,w)))
    else:
        ABO[(i,c,d,w)] >= 0,("Actual_BO_Chip" + str((i,c,d,w)))

for (m,c,d,w) in m_c_d_w_1:
    if w>=d and w<d+sigmam[(m,c)]:
        ABOM[(m,c,d,w)] >= BOM[(m,c,d,w)] - ALSM[(m,c,d,d)],("Actual_BO_Mod" + str((m,c,d,w)))
    else:
        ABOM[(m,c,d,w)] >= 0,("Actual_BO_Mod" + str((m,c,d,w)))
                                              
```

```python
#don't ship more than customers want
for (i,c,d,w) in i_c_d_w_1:
    problem += lpSum([ Q[(i,c,d,f,w)] for f in np.arange(1,F+1)]) <= DI[i,c,d,w],("Limit_Shipped_Chips" + str((i,c,d,w)))
for (m,c,d,w) in m_c_d_w_1:
    problem += lpSum([ QM[(m,c,d,f,w)] for f in np.arange(1,F+1)]) <= DM[m,c,d,w],("Limit_Shipped_Mods" + str((m,c,d,w)))
```

```python
#AVAILABLE CAPACITY

for (f,s,w) in f_s_w:
    problem += lpSum([X[(i,c,d,f,w)]* PTA[(i,f,s)] for (i,c,d) in i_c_d]) + lpSum([XM[(m,c,d,f,w)]* PTAM[(m,f,s)] for (m,c,d) in m_c_d]) <= CAA[(f,s,w)], ("Assem_Capacity" + str((f,s,w)))

for (f,t,w) in f_t_w:
    problem += lpSum([Y[(i,c,d,f,w)]* PTT[(i,f,t)] for (i,c,d) in i_c_d]) + lpSum([YM[(m,c,d,f,w)]* PTTM[(m,f,t)] for (m,c,d) in m_c_d]) <= CAT[(f,t,w)], ("Test_Capacity" + str((f,t,w)))

```

# Solve LP

```python
problem.solve()
```

```python
print("Total Cost = ", value(problem.objective))
```
